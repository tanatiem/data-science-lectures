```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
points = pd.read_csv('datasets/points.csv', header=None).values
new_points = pd.read_csv('datasets/new_points.csv', header=None).values
print('Train size: {}'.format(points.shape))
print('Test size: {}'.format(new_points.shape))
```

```python
model = KMeans(n_clusters=3)
model.fit(points)

labels = model.predict(new_points)
```

```python
print(labels)
```

### Train data

```python
plt.scatter(x=points[:,0], y=points[:,1], alpha=0.5);
```

### Test data and centroids

```python
x = new_points[:,0]
y = new_points[:,1]

plt.scatter(x, y, c=labels, alpha=0.5);

centroids = model.cluster_centers_
cx = centroids[:,0]
cy = centroids[:,1]
plt.scatter(cx, cy, marker='D', color='r', s=50);
```
