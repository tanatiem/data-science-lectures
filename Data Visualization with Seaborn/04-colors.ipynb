```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
college_df = pd.read_csv('datasets/college_datav3.csv')
print(college_df.shape)
college_df.head()
```

```python
college_df.info()
```

```python
sns.set()
tuitions = college_df['Tuition'].dropna()
```

# Color Palettes
Seaborn default palettes<br/>

```python
sns.palettes.SEABORN_PALETTES.keys()
```

```python
# f, axes = plt.subplots(3, 4, figsize=(16,8))
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    ax = sns.distplot(tuitions)
    ax.set_title('Pelettes: {}'.format(p))
    plt.show()
    
#     ax = axes[i]
#     ax = sns.distplot(tuitions, ax=ax)
#     ax.set_title('Palettes: {}'.format(p))
#     plt.show()
```

# Displaying Palettes

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```

### Circular colors
When data is not ordered

```python
sns.palplot(sns.color_palette('Paired',12))
```

### Sequential colors
When the daa has a consistent range from high to low

```python
sns.palplot(sns.color_palette('Blues',12))
```

### Diverging colors
When both the low and high values are interesting

```python
sns.palplot(sns.color_palette('coolwarm',12))
```
