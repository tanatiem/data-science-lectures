# Regression PLots

### Bicycle Dataset
- Aggregated bicycle sharing data in Washington DC
- Data includes:
 - Rental amounts
 - Weather information
 - Calendar information

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
bikedf = pd.read_csv('datasets/bike_share.csv', parse_dates=['dteday'])
bikedf.head()
```

## Linear regression
Temperature and Total Rentals

```python
sns.regplot(data=bikedf, x='temp', y='total_rentals', marker='+');
```

### Evaluate regression with residual plot
Ideally residual values should be plotted randomly across the horizontal line.<br>
The data looks like it might have a slight curve, suggesting a non-linear model might be appropriate.

```python
sns.residplot(data=bikedf, x='temp', y='total_rentals');
```

## Polynomial regression

```python
sns.regplot(data=bikedf, x='temp', y='total_rentals', order=2);
```

### Residual plot with polynomial regression
The residual values are more randomly distributed, so the 2nd order is likely more appropriate.

```python
sns.residplot(data=bikedf, x='temp', y='total_rentals', order=2);
```

## Categorical values

```python
sns.regplot(data=bikedf, x='mnth', y='total_rentals', order=2);
```

### Jitters

```python
sns.regplot(data=bikedf, x='mnth', y='total_rentals', x_jitter=.1, order=2);
```

### Estimators
This shows that the trend is consistent with seasons in Washington DC

```python
sns.regplot(data=bikedf, x='mnth', y='total_rentals', 
           x_estimator=np.mean, order=2);
```

## Bining the data
The regression line is still fit against all data

```python
sns.regplot(data=bikedf, x='temp', y='total_rentals', x_bins=4);
```

# College Dataset

```python
df = pd.read_csv('datasets/college_datav3.csv');
df.head()
```

### Regression plot

```python
# Display a regression plot for Tuition comparing to average SAT scores
plt.figure(figsize=(7,7))
sns.regplot(data=df, y='Tuition', x='SAT_AVG_ALL', marker='^', color='g', scatter_kws={'alpha':0.3});
```

### Residual plot

```python
# Display the residual plot
sns.residplot(data=df,
          y='Tuition',
          x="SAT_AVG_ALL",
          color='g');
```

## Regression plot with parameters

```python
plt.figure(figsize=(7,7))
sns.regplot(data=df, y='Tuition', x='REGION');
```

### Estimator

```python
plt.figure(figsize=(7,7))
sns.regplot(data=df, y='Tuition', x='REGION', x_estimator=np.mean);
```

### 3rd order polynomial

```python
plt.figure(figsize=(7,7))
sns.regplot(data=df, y='Tuition', x='REGION', x_estimator=np.mean, order=3);
```

## Binning Data

```python
sns.regplot(data=df, y='Tuition', x='UG', x_bins=5);
```

```python
sns.regplot(data=df, y='Tuition', x='UG', x_bins=8);
```
