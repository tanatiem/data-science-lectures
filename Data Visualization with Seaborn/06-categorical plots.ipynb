```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('datasets/schoolimprovement2010grants.csv', index_col=0)
df.head()
```

This is how the KDE distribution of the Award Amounts looks:

```python
sns.distplot(df['Award_Amount'], hist=False, kde_kws={'shade':True});
```

## Plot types: show individual observations

### Stripplot

```python
sns.stripplot(data=df, x='Award_Amount', y='Model Selected', jitter=True);
```

### Swarmplot

```python
plt.figure(figsize=(7,5))
sns.swarmplot(data=df, x='Award_Amount', y='Model Selected', hue='Region');
```

## Plot types: abstract representations

### Boxplot

```python
sns.boxplot(data=df, x='Award_Amount', y='Model Selected');
```

### Violinplot

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=df, x='Award_Amount', y='Model Selected', palette='husl');
```

### Boxenplot (lvplot)

```python
plt.figure(figsize=(10,8))
sns.boxenplot(data=df, x='Award_Amount', y='Model Selected', palette='Paired', hue='Region');
```

## Statistical Estimates
### Countplot

```python
sns.countplot(data=df,
             y='Model Selected',
             hue='Region');
```

## Pointplot

```python
# Create a pointplot and include the capsize in order to show bars on the confidence interval
sns.pointplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         capsize=.1);
```

### Barplot

```python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue='Region');
```
